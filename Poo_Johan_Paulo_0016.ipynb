##Lenguaje python, crear la clase "animal" con los atributos "nombre", "raza" crear constructor animal. y las funcions ladrar (), correr (), crear los objetos pitbull, chihuahua, nivel principiante, no es interactivo

```python
# Definición de la clase
class Animal:
    # Constructor
    def __init__(self, nombre, raza):
        self.nombre = nombre
        self.raza = raza

    # Función ladrar
    def ladrar(self):
        print(f"{self.nombre} está ladrando: ¡Guau guau!🐶")

    # Función correr
    def correr(self):
        print(f"{self.nombre} de raza {self.raza} está corriendo.👩‍🦽")

    def Morder(self):
        print(f"{self.nombre} está mordiendo a la veciana🦴")

# --- Creación de objetos ---

#1 Objeto Pitbull
pitbull = Animal("Rocco", "Pitbull")
pitbull.ladrar()
pitbull.correr()
pitbull.Morder()

print("-" * 20) # Separador visual

# Objeto Chihuahua
chihuahua = Animal("Chispa", "Chihuahua")

perrocallejero = Animal("solovino", "chafa")
chihuahua.ladrar()
chihuahua.correr()
chihuahua.Morder()
print("-" * 20) # Separador visual

perrocallejero.ladrar()
perrocallejero.correr()
perrocallejero.Morder()
print("Programa realizado por Johan Paulo NL = 04")
```
